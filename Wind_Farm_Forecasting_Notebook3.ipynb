## LSTM -  "Vanilla" - Single Hidden Layer - Daily Data
We will start our LSTM analysis on our daily average power outputs for the average of all four turbines in our daily_df.  We will start with 6 steps back or six lags so we can compare it to our modeling with random forest modeling. 

See Notebook 2, for an explanation of the process of using time lags to create independent variables and dependent variables for time series analysis. 

Through our research, LSTM models have had success analyzing time series using this approach. 

```python
# Dr. Brownlee's split_sequence function to create X and y for supervised learning
def split_sequence(sequence, n_steps):
    '''
    Frames a time series as a supervised learning dataset.
    Parameters
    ----------
        sequence: series 
        n_steps: Number of time lag observations as input (X).
    Returns
    -------
        np.array(X): Array of requested time steps or lags before dependent variable
        np.array(y): Array all dependent variables to make one step prediction for
    Examples
    --------
    >>>> dataframe columns or series of one timeseries signal, n_steps= 6,
    >>>> Two arrays: One with series for 6 time steps before each y or dependent variable
                    [[[ 305.9171875 ]
                      [ 228.81180556]
                      [ 608.49722222]
                      [1384.37829861]
                      [1198.80607639]
                      [  47.89305556]], ...
                      Another array of all the independent variables. One matching series above.
                      array([ 7.78088542e+01, ...
    '''
    X, y = list(), list()
    for i in range(len(sequence)):
        # find the end of this pattern
        end_ix = i + n_steps
        # check if we are beyond the sequence
        if end_ix > len(sequence)-1:
            break
        # gather input and output parts of the pattern
        seq_x, seq_y = sequence[i:end_ix], sequence[end_ix]
        X.append(seq_x)
        y.append(seq_y)
    return np.array(X), np.array(y)
```

```python
# choose a number of time steps
n_steps = 6
# split into samples
X, y = split_sequence(daily_df['four_turbine_avg_power'], n_steps)
# reshape from [samples, timesteps] into [samples, timesteps, features]
n_features = 1
X = X.reshape((X.shape[0], X.shape[1], n_features))
# define model
model1 = Sequential()
model1.add(LSTM(50, activation='relu', input_shape=(n_steps, n_features)))
model1.add(Dense(1))
model1.compile(optimizer='adam', loss='mse')
# fit model
model1.fit(X, y, epochs=200, verbose=0)
# demonstrate prediction
x_input = X
#x_input = x_input.reshape((X.shape[0], X.shape[1], n_features))
yhat = model1.predict(x_input, verbose=0)
print(x_input)
print(yhat)
```

We attempted to use Mean Absolute Percentage Error(MAPE)again. We are going to disregard this score as data values that are close to zero are making the MAPE overly inflated. We see this when we test training data with the hourly df as some hourly data is close to zero and the MAPE skyrockets. 

We will use median absolute error (med_ae) as a metric to see the value of the model disregarding extremely high power outputs and low power outputs.  Still, mean absolute error is going to be our best metric as our mean power output fluctuates widely and MAE encompasses.   

```python
 # estimate prediction error - 
# adding Median Absolute Error as a metric to see how outliers are weighing on the model
mae = mean_absolute_error(y, yhat)
r_square  = r2_score(y, yhat)
med_ae = median_absolute_error(y, yhat)
print(f'Median Absolute Error: {med_ae}')
print(f'Mean Absolute Error:{mae}')
print(f'R-squared:{r_square}')
```

```python
# Another function here to make a train and validation set from the input series
def train_val_split(time, series, time_step=342):
    '''
    Divide the time series into training and validation set
    Parameters
    ----------
        time: list of integers for time of each data point 
        series: series of data points for the time series analysis
        time_step: integer representing lenght of train set
    Returns
    -------
        time_train: list of integers representing times of train data points
        series_train: series odata points for the train set
        time_valid:  list of integers representing times of validation data points
        series_valid: series of data points for the validation set
    Examples
    --------
    >>>> list of times, series of data, time_step = 342,
    >>>> time_train = [0,1,2,...
    >>>> series_train = 0       305.917187
                        1       228.811806
                        2       608.497222...
    >>>> time_valid = [342, 343, 344,...
    >>>> series_valid = 342    1848.189757
                        343    1535.475000
                        344     792.990278...
    '''
    time_train = time[:time_step]
    series_train = series[:time_step]
    time_valid = time[time_step:]
    series_valid = series[time_step:]

    return time_train, series_train, time_valid, series_valid
```

```python
# Import all of  tensorlow as tf - changing naming of library to match functons
import tensorflow as tf
```

```python
# set default window_size to 6, batch_size to 16, shuffle_buffer size is size of dataset
def windowed_dataset(series, window_size=6, batch_size=16, shuffle_buffer=366):
    '''
    Creates time windows to make X and y features.
    
    Parameters
    ----------
        series: takes in series of data to be converted to X and y features
        window_size: Number of lag observations as input (X).
        batch_size: how many items will we supply per batch
        shuffle_buffer: this parameter to define the Tensorflow sample buffer
    Returns
    -------
       dataset: returns tensorflow.python.data.ops.dataset_ops.PrefetchDataset
    Examples
    --------
    >>>> time series data as series,  6, 16, 366
    >>>> tensorflow.python.data.ops.dataset_ops.PrefetchDataset 
        example output: [[ 302.04900651  298.25269116  172.7218272  1335.04547713 1082.84568956
                           1269.54052005]
                         [1269.54052005 1053.57788526  764.15188241  864.39497678  276.50185369
                           134.93385589]
    '''
    dataset = tf.data.Dataset.from_tensor_slices(series)
    dataset = dataset.window(window_size + 1, shift=1, drop_remainder=True)
    dataset = dataset.flat_map(lambda window: window.batch(window_size + 1))
    dataset = dataset.shuffle(shuffle_buffer)
    dataset = dataset.map(lambda window: (window[:-1], window[-1]))
    dataset = dataset.batch(batch_size).prefetch(1)
    return dataset
```

```python

def create_uncompiled_model():
    '''
    This funcion defines our model with all the tensorflow layers
    
    Parameters
    ----------
       None
    Returns
    -------
       LSTM Model with lambda layer to assure dimensionality is appropriate for
       five bidirectoinal layers
    '''
    model = tf.keras.models.Sequential([ 
        tf.keras.layers.Lambda(lambda x: tf.expand_dims(x, axis=-1),
                    input_shape=[None]),
        tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(1024, return_sequences=True)),
        tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(512, return_sequences=True)),
        tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(256, return_sequences=True)),
        tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(128, return_sequences=True)),
        tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(64)),
        tf.keras.layers.Dense(1),
    ]) 

    return model
```

```python
# This class allows the model to stop if a predefined metric is reached- 
#  we set our mae to less than 200 for early stopping
class EarlyStopping(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs={}):
        if(logs.get('mae') < 200):
            print("\nMAEthreshold reached. Training stopped.")
            self.model.stop_training = True

# Let's create an object of our class and assign it to a variable
early_stopping = EarlyStopping()
```

```python
# This function uses the create_uncompiled_model to 
def create_model():
    '''
    This funcion instantiates our model with our create_uncompiled_model().
    It then compiles the mdoel with our optimizer, loss, and scoring metric
    
    Parameters
    ----------
       None
    Returns
    -------
       Complete model to be later fitted
    '''
    # this reduces the random output of the model
    tf.random.set_seed(51)

    model = create_uncompiled_model()

    model.compile(loss=tf.keras.losses.Huber(), 
                  optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
                  metrics=["mae"])  # sets 
    return model
```

```python
#Instatiate model 
model2 = create_model()

# we fist our model for 20 epochs with and assign the callback of early stopping at MAE = 200
history = model2.fit(dataset, epochs=20, callbacks=[early_stopping])
```

```python
# Function to compute error metrics
def compute_metrics(true_series, forecast):
    '''
    This function takes in true values and predicted values and returns error scores.
    
    Parameters
    ----------
       true_series: series of true values
       forecast: series of predicted values
    Returns
    -------
       mse: mean squared error
       mae: mean absolute error
    '''
    mse = tf.keras.metrics.mean_squared_error(true_series, forecast).numpy()
    mae = tf.keras.metrics.mean_absolute_error(true_series, forecast).numpy()

    return mse, mae
#Function to return forecast  for visualization
def model_forecast(model, series, window_size):
    """"""
    '''
    This function converts the input series into a dataset with time windows for forecasting 
    and returns the predicted values for that series on the model.
    
    Parameters
    ----------
       model: model used (in this case the LSTM Bidrectional model = model2)
       series: series for forecast
       window_size: number of time lags in each window to set up our X variable
    Returns
    -------
       forecast: prediction for each value in series
    '''
    ds = tf.data.Dataset.from_tensor_slices(series)
    ds = ds.window(window_size, shift=1, drop_remainder=True)
    ds = ds.flat_map(lambda w: w.batch(window_size))
    ds = ds.batch(16).prefetch(1)
    forecast = model2.predict(ds)
    return forecast
```

```python
#We will run a for loop to make lists of n_steps and MAE and R2 Metrics to see which nstep
# between 12 and 40 n_steps
n_step_list = list(range(12,48))
r_square_list = []
mae_list = []
for n_step in n_step_list: # 20 epochs * 36 n_steps= 720 * 3sec - 5 sec = 36 - 60 minutes of run time
    # split into X and y 
    X, y = split_sequence(traindf['four_turbine_avg_power'], n_steps)
    # reshape from [samples, timesteps] into [samples, timesteps, features]
    X = X.reshape((X.shape[0], X.shape[1], n_features))
    # instantiate model
    model_hour = Sequential()
    model_hour.add(LSTM(50, activation='relu', input_shape=(n_steps, n_features)))
    model_hour.add(Dense(1))
    model_hour.compile(optimizer='adam', loss='mse', metrics=["mae"])
    # fit model
    model_hour.fit(X, y, epochs=20, verbose=1) #5 epochs to get an idea final test will run 33
    # demonstrate prediction
    x_input = X
    #x_input = x_input.reshape((X.shape[0], X.shape[1], n_features))
    yhat = model_hour.predict(x_input, verbose=0)
    # Metrics on each n_step
    mae = mean_absolute_error(y, yhat)
    r_square  = r2_score(y, yhat)
    # append scores to list
    mae_list.append(mae)
    r_square_list.append(r_square)
```

```python
#Fit model and predict on Train set - Later we will score test set
# We will run a full 200 epochs as we saw decrease in MAE all the way until the last 200 epoch 
# We will then adjust epoch numbers from there if we are landing on our best MAE. 
# choose a number of time steps
n_steps = 30
# split into samples
X, y = split_sequence(traindf['four_turbine_avg_power'], n_steps)
# reshape from [samples, timesteps] into [samples, timesteps, features]
n_features = 1
X = X.reshape((X.shape[0], X.shape[1], n_features))
# define model
model_hour = Sequential()
model_hour.add(LSTM(50, activation='relu', input_shape=(n_steps, n_features)))
model_hour.add(Dense(1))
model_hour.compile(optimizer='adam', loss='mse', metrics='mae')
# fit model
model_hour.fit(X, y, epochs=180, verbose=1) #171 * 3 sec = 8-9 minutes of run time
# demonstrate prediction
x_input = X
#x_input = x_input.reshape((X.shape[0], X.shape[1], n_features))
yhat = model_hour.predict(x_input, verbose=0)
print(x_input)
print(yhat)
```
